Create a House class that represents a house; attributes:


*   area
*   number of rooms
*   number of bathrooms
*   number of floors
*   has garage
*   has garden
*   any other attributes that you think is important



instance methods:

*   price calculation

There should be at least 1 default parameter

(4 points)

```python
class House:
  def __init__(self, street, house_num, area, num_of_rooms, num_of_bathrooms, num_of_floors, has_garage, has_garden, has_fence, area_of_garden, capable_of_living=True):
    self.street=street
    self.house_num=house_num
    self.area=area
    self.num_of_rooms=num_of_rooms
    self.num_of_bathrooms=num_of_bathrooms
    self.num_of_floors=num_of_floors
    self.has_garage=has_garage
    self.has_garden = has_garden
    self.has_fence=has_fence
    self.area_of_garden=area_of_garden
    self.capable_of_living=capable_of_living

  def price(self, price_per_sqr_h, price_per_sqr_g):
    #price_per_sqr_h=1500000
    #price_per_sqr_g=900000

    price=self.area*price_per_sqr_h

    if self.has_garden == True:
      price += self.area_of_garden*price_per_sqr_g

    if self.has_fence == True:
      price += 1200000

    if self.has_garage == True:
      price += 1650000

    return price


```

```python
class Animal:
  def __init__(self, fur_color, name, living_place):
    self.fur_color = fur_color
    self.name = name
    self.living_place = living_place

  def play(self):
    print(f"{self.name} likes to play.")

  def sound(self):
    print(f"{self.name} makes loud sounds.")

class Dog(Animal):
  def __init__(self, fur_color, name, living_place, breed):
    super().__init__(fur_color, name, living_place)
    self.breed = breed

  def sound(self):
    print(f"{self.name} makes loud sounds: woof")

  def fetch(self):
    print(f"{self.name} fetch the ball.")

class Cat(Animal):
  def __init__(self, fur_color, name, living_place, jump_height):
    super().__init__(fur_color, name, living_place)
    self.jump_height = jump_height

  def sound(self):
    print(f"{self.name} makes quite sounds: Meow")

  def scratching(self):
    print(f"{self.name} is scratching the scratching post.")

class Hamster(Animal):
  def __init__(self, fur_color, name, living_place, wheel_size):
    super().__init__(fur_color, name, living_place)
    self.wheel_size = wheel_size

  def sound(self):
    print(f"{self.name} makes quite sounds: squeak")

  def rolling(self):
    print(f"{self.name} is rolling the wheels.")

dog = Dog("black", "Rex", "garden", "German Shepherd")
cat = Cat("grey", "Cirmi", "inside", "2 meters")
hamster = Hamster("brown", "Cucu", 'cage', "22 cm")

dog.play()
dog.sound()
cat.sound()
cat.scratching()
hamster.play()
hamster.rolling()

```

Create classes and use inheritance to calculate the area of different shapes (specifically, circle and square, you can add triangle or anything else if you want).

Requirements:


1.	Create a base class called Shape that has a method named area().

This method should not have any implementation

2. Create at least two subclasses:

Circle — which takes a radius as a parameter

Square — which takes a side as a parameter

write the area-calculating function for both (or for all)

3. Create objects of both Circle and Square, and print their areas.

Extra: Store these objects in a list and use a loop to print the area of each shape.

```python
import math as m

class Shapes:
  def area(self):
    pass

class Circle(Shapes):
  def __init__(self, radius):
    self.radius=radius

  def area(self):
    y = m.pi * self.radius ** 2
    print(f"The area of the circle: {y}.")


class Square(Shapes):
  def __init__(self, a):
    self.a=a

  def area(self):
    t = self.a**2
    print(f"The area of the square: {t}.")

class Rectangle(Shapes):
  def __init__(self, a, b):
    self.a=a
    self.b=b

  def area(self):
    x = self.a*self.b
    print(f"The area of the rectangle: {x}.")

C1 = Circle(3)
C2 = Circle(5)
S1 = Square(4)
S2 = Square(6)
R1 = Rectangle(2,4)
R2 = Rectangle(5,7)

C1.area()
C2.area()
print()
S1.area()
S2.area()
print()
R1.area()
R2.area()

```
